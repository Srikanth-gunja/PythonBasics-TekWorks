```python
# Min 10 ,max of 15
def password_validator(password):
    if  len(password) < 10 or len(password) >15:
        return "password should min 10 and below 15"
    elif not any(c.isalnum() for c in password):
        return "password contains digits and letter"
    elif not any(c.isupper() for c in password):
        return "Not capital letter"
    elif not any(c in ['@','#','$'] for c in password):
        return "No Special Character"
    elif not len(password) == len(password.strip()):
        return "No white spaces"
    else:
        return "Valid"
print(password_validator("Yyyyyohh@yg"))
```

```python
# Factorail
def fact(n):
    if n<=1:
        return 1
    else:
        return n*fact(n-1)
print(fact(5))
```

```python
def calculate_total(cart):
  
    
    total=sum(cart.values())
    if(total> 20000 or total < 50000):
            total *= 0.1
    elif total>50000:
            total*=0.15
    return total
cart={
    'Laptop':50000,'Headphones':2000,'Mouse':35000,'Keyboard':1500,'Monitor':8000,'USB drive':1000
}
print(f"The discount is {calculate_total(cart)}")
```
